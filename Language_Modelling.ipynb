Write a parser that can read all the lines of the file `holbrook.txt` and print out for each line the original (misspelled) text, the corrected text and the indexes of any changes. The indexes refers to the index of the words in the sentence. In the example given, there is only an error in the 10th word and so the list of indexes is [9]. It is not necessary to analyze where the error occurs inside the word.

```
import nltk
nltk.download('punkt')
def parse(sent):
  lorig = []
  lcorrect = []
  indexes = []
  cnt = 0
  for i in sent:
        if '|' in i:
          # Splitting the sentence on '|'
            spt = i.split('|')
            # Previous word to '|' is storing in lorig list.
            lorig.append(spt[0])
            # Next word to '|' is storing in lcorrect list.
            lcorrect.append(spt[1])
            #Noting down the index of error.
            indexes.append(cnt)
        else:
            # If there is no '|' in sentence, sentence is stored in loriginal and lcorrected as it is.
            lorig.append(i)
            lcorrect.append(i)
        cnt = cnt + 1
  #Loading to lorig, lcorrect and index list to dictionary.      
  dictionary = {'corrected': lcorrect, 'original': lorig, 'indexes': indexes}
  return dictionary


lines = open("holbrook.txt").readlines()
print(len(lines))
data = []
# Write your code here

# Reading the txt file in reading mode
corp_read = open("holbrook.txt", "r")
lines = []
for i in corp_read:
  lines.append(i.strip())

print(lines[1])

# Tokenizing the sentences based on Words
word_tok = [nltk.word_tokenize(i) for i in lines]
  
for i in word_tok:
  data.append(parse(i))
#Printing the Data
for i in data:
  print i



#assert(data[2] == {
 #'original': ['I', 'have', 'four', 'in', 'my', 'Family', 'Dad', 'Mum', 'and', 'siter', '.'], 
  #'corrected': ['I', 'have', 'four', 'in', 'my', 'Family', 'Dad', 'Mum', 'and', 'sister', '.'], 
  #'indexes': [9]
#})
```

## **Task 3** (15 Marks): 
[Edit distance](https://en.wikipedia.org/wiki/Edit_distance) is a method that calculates how similar two strings are to one another by counting the minimum number of operations required to transform one string into the other. There is a built-in implementation in NLTK that works as follows:


```
from nltk.metrics.distance import edit_distance

# Edit distance returns the number of changes to transform one word to another
print(edit_distance("hello", "hi"))
```

Write a function that calculates all words with *minimal* edit distance to the misspelled word. You should do this as follows

1. Collect the set of all unique tokens in `train`
2. Find the minimal edit distance, that is the lowest value for the function `edit_distance` between `token` and a word in `train`
3. Output all unique words in `train` that have this same (minimal) `edit_distance` value

*Do not implement edit distance, use the built-in NLTK function `edit_distance`*

```
from nltk.metrics.distance import edit_distance
nltk.download('averaged_perceptron_tagger')
def get_candidates(tok):
    
    corrList = []
    #Create a list of the corrected training data
    for i in train_correct:
        corrList.append(" ".join(i))

    corrList = " ".join(corrList)
    token = nltk.word_tokenize(corrList)
    # the frequency of each vocabulary item in the text
    fd = FreqDist(token)
    # unique words this corpus have
    unique_freq = fd.keys()
    
    # Calculate distance between the given word and all words in corpus
    s = []
    for i in unique_freq:
        
        t = edit_distance(i,tok)
        
        s.append(t)
    
    # Create a key Value Pair of the distance and the words in the corpus
    dist = dict(zip(unique_freq, s))
    
    dist_sort = sorted(dist.items(), key=lambda x: x[1])
    #print("Sorted Tuples of words :",dist_sort)
    #Find the minimum distance
    k,v = dist_sort[0]
    minimal_dist = v
    keys_min = filter(lambda d: d[1] == minimal_dist, dist_sort)
    final_keys = [x[0] for x in keys_min]
    
    return final_keys
  
#print("The words closest to the passed word using the edit distance function are", get_candidates("minde"))

 
   # return []
res = get_candidates('minde')
print(res)
# Test your code as follows
#assert get_candidates("minde") == ['mind', 'mine']
```

## Task 4 (15 Marks):

Write a function that takes a (misspelled) sentence and returns the corrected version of that sentence. The system should scan the sentence for words that are not in the dictionary (set of unique words in the training set) and for each word that is not in the dictionary choose a word in the dictionary that has minimal edit distance and has the highest *unigram probability*. 

*Your solution to this should involve `get_candidates`*


```
corrList = []
#Create a list of the corrected training data
test_corrList = []
for i in test_correct[42:45]:
       test_corrList.append(" ".join(i).lower())
test_corrList = " ".join(test_corrList)

print (test_corrList)
for i in train_correct:
  corrList.append(" ".join(i))

corrList = " ".join(corrList)
token = nltk.word_tokenize(corrList)
# the frequency of each vocabulary item in the text
fd = FreqDist(token)
# unique words this corpus have
unique_freq = fd.keys()

def correct(sentence):
    # Write your code here
    corrected = []
    
    indexes = []
    for i in sentence:
        # If word not in unique words then calculate suggested words with get_candidate() function
        
        if i.lower() not in unique_freq:
          #indexes.append(cnt)
          
          l = len(get_candidates(i))
          
          if l > 1:
                suggestion = get_candidates(i)
                probability = []
                
                # For each suggested word calculate unigram probability
                for sug in suggestion:
                  p = prob(sug)
                  probability.append(p)
                
                #The suggested word with maximum probability
                corrected.append(suggestion[probability.index(max(probability))])
          else:
            # If there is only 1 suggested word then take that word - no need to calculate probabilities
            corrected.append(get_candidates(i)[0])
        else:
          corrected.append(i)
          
    return corrected

correct(['My', 'siter', 'go', 'to', 'Tonbury'])

#assert(correct(["this","whitr","cat"]) = ['this','white','cat'])   
```

## **Task 5** (10 Marks): 
Using the test corpus evaluate the *accuracy* of your method, i.e., how many words from your system's output match the corrected sentence (you should count words that are already spelled correctly and not changed by the system).

```
def accuracy():
    # Write your code here
    
    class_test_corr = []
    #Create a list of the corrected test data using the classifier created above
    for i in test_original:
      t = correct(i)
      class_test_corr.append(t)
    total = 0
    count = 0
    sent_corr = []
    #Iterating over test Correct Data
    for i in test_correct:
   
        t = len(i)
        #print("lenght" ,t)
        total = 0
        #If there is no words present in a sentence then its an empty list and hence assigning 100% accuracy for that sentence
        if (t == 0):
          sent_corr.append(1)
          count = count +1
        else:
          for j in range(0,len(class_test_corr[count])):
            #If the output of the classifier contains the word present in the Correct Test Set then we add up the total words correctly predicted by classifier
            if (class_test_corr[count][j] in i):
              total = total + 1
              
          #Accuracy is defined as the total no of correct words predicted by the classifier / Total no of words in that sentence.
          acc = total/t
          sent_corr.append(acc)
          count = count +1
          
    #Averaging the accuracy over all the sentences in the Train Corpus.
    print("Average Accuracy of words in each sentence: " , round(sum(sent_corr)/len(sent_corr)*100,4), "%")
    
    

#print(accuracy(test))
accuracy()
```
